## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
df=pd.read_csv('datasets/Countries-exercise.csv')
```

```python
df.sample(5)
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(df['Longitude'],df['Latitude'])
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
x=df.iloc[:,1:3]
x.sample(5)
```

## Clustering

Assume there are only two clusters. 

```python
km=KMeans(2)
```

```python
km.fit(x)
```

### Clustering Resutls

```python
indentified_cluster=km.predict(x)
indentified_cluster
```

```python
data_with_cluster=df.copy()
data_with_cluster['Cluster']=km.predict(x)
data_with_cluster.sample(5)
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
plt.scatter(df['Longitude'],df['Latitude'],c=data_with_cluster['Cluster'],cmap='rainbow')
plt.show()
```
