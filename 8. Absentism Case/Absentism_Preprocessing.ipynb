## PreProcessing

### Importing Libraries

```python
import pandas as pd
```

```python
raw_csv_data=pd.read_csv('Absenteeism_data.csv')
```

```python
raw_csv_data.sample(5)
```

### Dropping ID

```python
raw_csv_data=raw_csv_data.drop('ID',axis=1)
```

### Grouping Reason For Absenties

```python

raw_csv_data['Reason 1']=0
raw_csv_data['Reason 2']=0
raw_csv_data['Reason 3']=0
raw_csv_data['Reason 4']=0
    
for i in range(len(raw_csv_data)):
    if raw_csv_data['Reason for Absence'][i] in range(1,15):
        raw_csv_data['Reason 1'][i]=1
    elif raw_csv_data['Reason for Absence'][i] in range(15,18):
        raw_csv_data['Reason 2'][i]=1
    elif raw_csv_data['Reason for Absence'][i] in range(18,22):
        raw_csv_data['Reason 3'][i]=1
    else:
        raw_csv_data['Reason 4'][i]=1
```

```python
raw_csv_data=raw_csv_data.drop('Reason for Absence',axis=1)
```

### Extracting Month Value and Day of Week from Date Column

```python
raw_csv_data['Month Value']=0
raw_csv_data['Day of the Week']=0

for i in range(len(raw_csv_data)):
    raw_csv_data['Day of the Week'][i]=raw_csv_data['Date'][i].split('/')[0]
    if raw_csv_data['Day of the Week'][i] in range(1,8):
        raw_csv_data['Day of the Week'] = 1
    if raw_csv_data['Day of the Week'][i] in range (8,15):
        raw_csv_data['Day of the Week'] = 2
    if raw_csv_data['Day of the Week'][i] in range (15,22):
        raw_csv_data['Day of the Week'] = 3
    else:
        raw_csv_data['Day of the Week'] = 4
    raw_csv_data['Month Value'][i]=raw_csv_data['Date'][i].split('/')[1]
```

```python
raw_csv_data
```

```python
raw_csv_data=raw_csv_data.drop('Date',axis=1)
```

### Mapping Education

```python
raw_csv_data['Education']=raw_csv_data['Education'].map({1:0,2:1,3:1,4:1})
```
